```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X= np.array([
    [0,1],
    [3,0],
    [4,0],
    [6,0],
    [3,0],
    [4,0],
    [0,1],
    [6,0],
])
# model=KMeans(n_clusters=4,random_state=42)
# model.fit(X)

# clusters=model.predict(X)
# plt.scatter(x[:,0],x[:,1],c=clusters, marker='o',cmap='viridis')
```

```python
# Step 1: Apply Different Clustering Algorithms
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
 
#1. K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
 
#2. DBSCAN Clustering
dbscan = DBSCAN(eps=1.5, min_samples=2)
dbscan_labels = dbscan.fit_predict(X)
 
#3. Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=4)
agglo_labels = agglo.fit_predict(X)
 
 
#Step 2: Evaluate Clustering Results
 
#Silhouette Score
kmeans_silhouette = silhouette_score(X, kmeans_labels)
dbscan_silhouette = silhouette_score(X, dbscan_labels)
agglo_silhouette = silhouette_score(X, agglo_labels)
 
#Davies-Bouldin Index
kmeans_db = davies_bouldin_score(X, kmeans_labels)
dbscan_db = davies_bouldin_score(X, dbscan_labels)
agglo_db = davies_bouldin_score (X, agglo_labels)
```

```python
print("K-Means Silhouette Score:", kmeans_silhouette, " Davies-Bouldin Score:", kmeans_db)
print("DBSCAN Silhouette Score:", dbscan_silhouette," Davies-Bouldin Score:", dbscan_db)
print("Agglomerative Silhouette Score:", agglo_silhouette," Davies-Bouldin Score:", agglo_db)
 
#Step 3: Fine-Tune the Best-Performing Model (Example with K-Means)
best_model = kmeans #Assuming K-Means performs best
best_labels = kmeans_labels
 

```

```python
#Visualizing the best clustering result
 
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=best_labels, cmap='viridis', marker='o')
plt.title('Best Shape Clusters')
plt.xlabel('Sides')
plt.ylabel('Curvature (1: Curved, 0: Not Curved)')
plt.colorbar(label='Cluster Label')
plt.show()
 
```

```python
#Displaying the final cluster labels for the original shapes 
for i, cluster in enumerate (best_labels):
 

    print(f"Shape: {X[i]}, Cluster: {cluster}")
```
