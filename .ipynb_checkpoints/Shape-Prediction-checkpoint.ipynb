import cv2
import numpy as np
import pandas as pd
import os
from tqdm import tqdm

# Function to calculate aspect ratio (width/height)
def calculate_aspect_ratio(contour):
    x, y, w, h = cv2.boundingRect(contour)
    return w / h

# Function to calculate compactness (Perimeter^2 / (4 * pi * Area))
def calculate_compactness(area, perimeter):
    if area == 0: # Prevent division by zero
        return 0
    return (perimeter ** 2) / (4 * np.pi * area)

# Function to calculate solidity (Area / Convex Hull Area)
def calculate_solidity(contour, area):
    hull = cv2.convexHull(contour)
    hull_area = cv2.contourArea(hull)
    if hull_area == 0: # Prevent division by zero
        return 0
    return area / hull_area

# Function to calculate convexity (Convex Hull Perimeter / Shape Perimeter)
def calculate_convexity(contour, perimeter):
    hull = cv2.convexHull(contour)
    hull_perimeter = cv2.arcLength(hull, True)

    if perimeter == 0: # Prevent division by zero
        return 0
    return hull_perimeter / perimeter

# Function to detect and classify shapes based on contours
def detect_shape(contour, known_shape=None):
# Approximate the contour to reduce the number of points
    approx = cv2.approxPolyDP(contour, 0.04 * cv2.arcLength(contour, True), True)
    sides = len(approx)
    # If a known shape is provided (from folder), use that as the label
    if known_shape and known_shape=='circle':
        return "Circle",sides
    if known_shape and known_shape=='square':
        return "Square",sides
    if known_shape and known_shape=='triangle':
        return "Triangle",sides
    if known_shape and known_shape=='star':
        return "Star",sides


# Load and process images from directories

def process_images_from_folders(base_dir):
    data = []
    # Iterate through each folder (e.g., square, circle, triangle, star)
    for folder_name in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, folder_name)
        if os.path.isdir(folder_path): # Ensure it's a directory
            for flename in tqdm(os.listdir(folder_path), desc=f"Processing {folder_name}", leave=True):
                if flename.endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(folder_path, flename)
                    # Load the image and convert to grayscale
                    image = cv2.imread(image_path)
                    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                    # Apply threshold to get binary image
                    _, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)
                    # Find contours in the image
                    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                    for contour in contours[0:1]:
                        # Calculate area and perimeter
                        area = cv2.contourArea(contour)
                        perimeter = cv2.arcLength(contour, True)
                        # Detect the shape and get the number of sides
                        shape_name, sides = detect_shape(contour, known_shape=folder_name)
                        # Calculate additional features
                        aspect_ratio = calculate_aspect_ratio(contour)
                        compactness = calculate_compactness(area, perimeter)
    
                        solidity = calculate_solidity(contour, area)
                        convexity = calculate_convexity(contour, perimeter)
                        # Add data to the list
                        data.append([
                        flename, shape_name, sides, area,
                        perimeter, aspect_ratio, compactness,
                        solidity, convexity
                        ])
    # Create a DataFrame from the data
    df = pd.DataFrame(data, columns=[
    "Image Name", "Shape Name", "Sides", "Area",
    "Perimeter", "Aspect Ratio", "Compactness",
    "Solidity", "Convexity"
    ])
    return df

# Defne the base directory containing your shape folders
base_directory = "./shapes/"

# Process the images and create the dataset
shapes_df = process_images_from_folders(base_directory)

# Display the dataset
print(shapes_df)

shapes_df.to_csv("shapes_dataset_with_additional_features.csv", index=False)
