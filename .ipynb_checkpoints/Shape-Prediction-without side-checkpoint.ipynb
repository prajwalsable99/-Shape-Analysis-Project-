```python
from sklearn.cluster import KMeans





#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_train)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
#instantiate the k-means class, using optimal number of clusters
kmeans = KMeans(init="random", n_clusters=4, n_init=10, random_state=1)

#fit k-means algorithm to data
kmeans.fit(x)

#view cluster assignments for each observation
kmeans.labels_
```

```python
def mapit(x):
    if(x==2):
        return "Triangle"
    if(x==3):
        return "Square"
    if(x==1):
        return "Circle"
    if(x==0):
        return "Star"
```

```python
kmeansdf['predicted cluster mapped class']=kmeansdf['predicted cluster'].apply(mapit)
```

```python
kmeansdf
```

```python
confusion_matrix(kmeansdf['original labels'],kmeansdf['predicted cluster mapped class'])
```
